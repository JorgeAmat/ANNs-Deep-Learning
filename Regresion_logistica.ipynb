```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Pesos (X)
# Â¿Es perro o no? (T)

x = np.array([-0.13459237, -3.3015387, 0.74481176, 2.62434536, 0.38824359, 0.47182825, -0.07296862]).reshape(-1, 1)

t = np.array([0, 0, 0, 1, 1, 1, 1]).reshape(-1, 1)

cat_mask = (t == 0).ravel()
dog_mask = (t == 1).ravel()

plt.figure(figsize=(7,5))

plt.scatter(x[cat_mask], t[cat_mask], color='C0', marker='o', label='data observations class 0 cat')

plt.scatter(x[dog_mask], t[dog_mask], color='C1', marker='*', s=150, label='data observations class 1 dog')

plt.xlabel('peso')
plt.ylabel('probability of being dog')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

linear_models= [
    {"w": 2.0,  "b": -0.2, "label": "Modelo A"},
    {"w": 4.0,  "b": -1.0, "label": "Modelo B"},
    {"w": -3.0, "b":  0.8, "label": "Modelo C"},
]

x_line = np.linspace(x.min() - 0.5, x.max() + 0.5, 400)

plt.figure(figsize=(8,5))

plt.scatter(x[cat_mask], t[cat_mask], color='C0', marker='o', label='class 0 cat')
plt.scatter(x[dog_mask], t[dog_mask], color='C1', marker='*', s=150, label='class 1 dog')

for m in linear_models:
    y_line = sigmoid(m["w"] * x_line + m["b"])
    plt.plot(x_line, y_line, linewidth=2, label=f'{m["label"]} (w={m["w"]}, b={m["b"]})')

plt.xlabel('peso')
plt.ylabel('p(t=1 | x)  =  probabilidad de ser perro')
plt.ylim(-0.1, 1.1)
plt.legend()
plt.grid(True)
plt.title('2.2 Posibles modelos lineales (logistic) + datos')
plt.show()
```
